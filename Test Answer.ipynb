```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("used_car_data.csv")
```

```python
df.head()
```

# Nomor 1

```python
dfbaru = pd.DataFrame()
```

```python
a = 0
for x in df['Name']:
    if 'Land' in x:
        b = x.split(' ',2)[:2]
        dfbaru.at[a,'Name'] = b[0]+' '+b[1]
        a = a+1
    else:
        dfbaru.at[a,'Name'] = x.split(' ',1)[0]
        a = a+1
```

```python
dfbaru
```

```python
dfbaru['Name'] = dfbaru['Name'].str.lower().str.capitalize()
```

```python
dfbaru
```

```python
Merk = dfbaru.pivot_table(index='Name',aggfunc="size").sort_values(ascending=False)
```

```python
No1 = pd.DataFrame({'Name':Merk.index,'Jumlah':Merk.values})
```

```python
No1
```

```python
sns.set(style="whitegrid")
f, ax = plt.subplots(figsize=(15,6))
sns.barplot(x='Jumlah',y='Name',data=No1,color='b')
```
